```python
from math import *
import cmath

#fonction pour calculer l'hypotenus 
def calculer(a,b):
     c =  sqrt(a**2 + b**2)
     print('l hypotenus formé par' ,a, ' et ' ,b, ' vaut : ',c)
     return c



# fonction pour vérifier  un nombre et le limiter
def si_nombre(a):
  limite = range(-999,999) # liste pour limiter les inputs 
  compteur = 0
  while(compteur <=3 ): # boucle afin de réitérer l'opération 
     a = int(a)
     if(a>0 and a in limite): # verifier en cas d'entier positif 
             return a
             break 
     elif(a<0 and a in limite): # vérifier en cas d'entier négatif 
             print('la valeur de ',a,' est negative ,  poursuivre ? ')
             retour = int(input(' si oui tapez 1 sinon 0 ') )
             if (retour == 0): # changer le nombre 
                   a = input('nouvelle valeur : ' )
                   compteur +=1
             elif (retour == 1): # poursuivre le calcul malgré l'entier négatif 
                   a = abs(a) 
                   return a
                   break
     elif(a not in limite):  # en cas de caractère long 
            print('le caractère ',a,' est trop long')  
            a = input(' entrez un nouveau entier  ')
            compteur  +=1    


# fonction pour vérifier  un complexe  et le limiter
def si_complexe(a):
  limite = range(-999,999) # liste pour limiter les inputs 
  compteur = 0
  while(compteur <= 3 ):  # boucle afin de réitérer l'opération 
      a  = complex(a)
      if(a.real > 0  and a.real in limite  ): # verifier en cas d'entier positif
             return a.real
             break
      elif(a.real < 0  and a.real in limite    ):
             print('la partie réelle ',a.real,' est negative ,  poursuivre ? ')
             retour = int(input(' si oui tapez 1 sinon 0 ') )
             if (retour == 0):   # changer le nombre 
                   a = input('nouvelle valeur : ' )
                   compteur +=1
             elif (retour == 1): # poursuivre le calcul malgré l'entier négatif 
                   return abs(a.real)               
                   break
      elif(a.real  not in limite  ): # en cas de caractère long 
            print('le caractère ',a,' est trop long')  
            a = input(' entrez  un nouveau complexe  ')
            compteur  +=1 
     


# fonction  pour reprendre en cas de caractère string 
def recommencer(a):
  compteur = 0
  while(compteur <= 4 ): # boucle afin d'itérer l'opération 
      if(a.isalpha()):  # tester s'il s'agit d'un str
            print('le caractère ', a ,'ne correspond pas '  )
            a = input('mettre un nombre ou un complexe vous avez 8 essais  : ')
            compteur +=1  
      else: # sinon 
           try : # essayer de voir si c'est un entier 
                a = int(a) 
                return a  
           except ValueError: # sinon éssayer de voir si c'est un entier 
                a = complex(a)  
                return a
           except ValueError: # sinon remettre de nouveau input 
                print('format impossible')  
                compteur +=1


#fonction pour tester les caractères 
def tester_le_type(a,b):           
  if(type(a) == int and type(b) == int): # condition  et opération au cas ou on a 2 entiers 
        a = si_nombre(a)  # faire les opérations sur ce 1  input 
        b = si_nombre(b)  # faire les opérations néccésaires sur le 2eme input 
        n = calculer(a,b) # obtenir l'hypotenuse
  elif(type(a) == complex and type(b) == complex): # conditon et opération au cas ou on a 2   complexes
        a = si_complexe(a) 
        b = si_complexe(b)
        n = calculer(a,b) 
  if(type(a) == int and type(b) == complex): # condition et opération au cas ou on a un entier et complexe 
        a = si_nombre(a) 
        b = si_complexe(b)
        n = calculer(a,b)
  elif(type(a) == complex and type(b) == int): # condition et opération au cas ou on a un complexe et un entier 
        a = si_complexe(a) 
        b = si_nombre(b)
        n = calculer(a,b)
                      


 # fonction pythagore qui retournera les resultats en fonction des inputs                
def pythagore():
   a = input('entrer une valeur pour a : ' )
   b = input('entrer une valeur pour b : ' )
   if (a.isalpha() and b.isalpha()) and (a.isalpha() or b.isalpha()): # résultat retourné si a et b sont des str
         a = recommencer(a)
         b = recommencer(b)
         tester_le_type(a,b)
   else : # dans le cas ou les inputs ne sont des str 
     try : # résultat retourné si  a et b sont des entiers 
         a,b = int(a), int(b)
         tester_le_type(a,b)
     except ValueError: # résultat retourné si a et b sont des complexes
         a,b = complex(a), complex(b)
         tester_le_type(a,b)
     except ValueError: # résultat retourné si a est un entier et b un  complexe
         a,b = int(a), complex(b)
         tester_le_type(a,b)
     except ValueError: # recommencer l'opération dans le cas contraire
         a , b = recommencer(a),recommencer(b)
            
            
pythagore()    # rendu final         
```
